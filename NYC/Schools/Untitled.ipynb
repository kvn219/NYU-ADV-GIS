```python
import pandas as pd
import numpy as np
import geopandas as gp
import csv
import shapefile
import re

pd.options.display.max_rows = 999
pd.options.display.max_columns = 500
```

```python
geo_file = pd.read_csv("./merged_geojson.csv")
geo_file.head()
```

```python
geo_file['% Poverty'] = [float(re.sub('[%]', '', rate)) for rate in geo_file['% Poverty']]
geo_file['% Students with Disabilities'] = [float(re.sub('[%]', '', rate)) for rate in geo_file['% Students with Disabilities']]
geo_file['% English Language Learners'] = [float(re.sub('[%]', '', rate)) for rate in geo_file['% English Language Learners']]
```

```python
geo_file_to_test = geo_file[['@id', 'schoolDistrict', '% Poverty', '% Students with Disabilities', '% English Language Learners', 'geometry']]
geo_file_to_test.columns = ['id', 'district', 'poverty', 'disabilities', 'english_language_learners', 'geometry']
geo_file_to_test.to_csv("test.csv", encoding='utf-8', index=False)
geo_file_to_test.head()
```

```python
# Create a polygon shapefile writer
w = shapefile.Writer(shapefile.POLYGON)

# Add our fields
w.field("id", "C", "60")
w.field("district", "N", "40")
w.field("poverty", "N", "40")
w.field("disabilities", "N", "40")
w.field("english_language_learners", "N", "40")

# Open the csv file and set up a reader
with open("test.csv") as p:
    reader = csv.DictReader(p)
    for row in reader:
        # Add records for each polygon for name and area
        w.record(row["id"], row["district"], row["poverty"], row["disabilities"], row["english_language_learners"])
        # parse the coordinate string
        wkt = row["geometry"][9:-2]
        # break the coordinate string in to x,y values
        coords = wkt.split(",")
        # set up a list to contain the coordinates
        part = []
        # convert the x,y values to floats
        for c in coords:
            c = re.sub('[()]', '', c)
            x,y = c.strip().split(" ")
            part.append([float(x),float(y)])
        # create a polygon record with the list of coordinates.
        w.poly(parts=[part])

# save the shapefile!
w.save("./shapefile/nyc_school_districts.shp")
```

```python
import pandas as pd
import geopandas as gpd

import fiona.crs
from shapely.geometry import Point

# Open the file as a table. Name the two columns as x and y
df = pd.read_csv('raw.txt', header=None, names=['x', 'y'], sep=' ')
# Convert to shapely Points and to a GeoDataFrame with epsg 26919
geoms = df.apply(lambda row: Point(row['x'], row['y']), axis=1)
gdf = gpd.GeoDataFrame({'geometry': geoms}, crs=fiona.crs.from_epsg(26919))
# Convert to lon/lat
gdf.to_crs(epsg=4326, inplace=True)
# Store GeoJSON
with open('converted.geojson', 'w') as f:
    f.write(gdf.to_json())
```
